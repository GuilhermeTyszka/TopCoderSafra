```python
import json
import os.path
import sys

def obter_dados():
    '''
    Essa função carrega os dados dos produtos e retorna uma lista de dicionários, onde cada dicionário representa um produto.
    NÃO MODIFIQUE essa função.
    '''
    with open(os.path.join(sys.path[0], 'dados.json'), 'r') as arq:
        dados = json.loads(arq.read())
    return dados

def listar_categorias(dados):
    categorias_set = set([])
    for i in dados:
        categorias_set.add(i['categoria'])
    return list(categorias_set)

def escolher_categoria(dados):
    categoria = input('Digite a categoria que deseja listar: ')
    voltar_menu = True
    while categoria not in listar_categorias(dados) and voltar_menu:
        categoria = input('Esta categoria não existe, por favor digite uma categoria válida ou digite 9 para voltar ao menu anterior!\n')
        if categoria == '9': voltar_menu = False
    return categoria


def listar_por_categoria(dados, categoria):
    produtos_list = []
    if categoria in listar_categorias(dados):
        for i in dados:
            if i['categoria'] == categoria:
                produtos_list.append(i)
    else: 
        print('Categoria não encontrada!')
    return produtos_list         
    

def produto_mais_caro(dados, categoria):
    return sorted(listar_por_categoria(dados, categoria), key = lambda x: float(x['preco']))[-1]
    
def produto_mais_barato(dados, categoria):
    return sorted(listar_por_categoria(dados, categoria), key = lambda x: float(x['preco']))[0]

def mais_caro(dados):
    return sorted(dados, key = lambda x: float(x['preco']))[-1] 

def mais_barato(dados):
    return sorted(dados, key = lambda x: float(x['preco']))[0] 

def top_10_caros(dados):
    return sorted(dados, key = lambda x: float(x['preco']))[-11:]
    
def top_10_baratos(dados):
    return sorted(dados, key = lambda x: float(x['preco']))[:10]
    
def imprimir_dados(dados):
    for i in dados:
        print(f'Produto: ID: {i["id"]}, categoria: {i["categoria"]} e preço R$ {i["preco"]}.')

def imprimir_dados_tabela(dados):
    col_1 = len(sorted(dados, key = lambda x: len(x['id']))[-1]['id'])
    col_2 = len(sorted(dados, key = lambda x: len(x['categoria']))[-1]['categoria'])
    col_3 = len(sorted(dados, key = lambda x: len(x['preco']))[-1]['preco'])
    print('+' + '-'*(col_1 + col_2 + col_3 + 13) + '+')
    print('|',' '*((col_1 - len('ID'))//2), 'ID',' '*((col_1 - len('ID'))//2-1),'|', ' '*((col_2 - len('CATEGORIA'))//2), 'CATEGORIA', ' '*((col_2 - len('CATEGORIA'))//2 + 1), '|', ' '*((col_3 - len('preco'))//2), 'PREÇO', ' '*((col_3 - len('preco'))//2), '|')
    print('+' + '-'*(col_1 + col_2 + col_3 + 13) + '+')
    for i in dados:
        espacos_col_2_esq =  (col_2 - len(i['categoria']))//2
        espacos_col_2_dir = (col_2 - len(i['categoria']))//2 if ((col_2 - len(i['categoria'])) % 2 == 0) else (col_2 - len(i['categoria']))//2 + 1
        espacos_col_3_esq =  (col_3 - len(i['preco']))//2
        espacos_col_3_dir = (col_3 - len(i['preco']))//2 if ((col_3 - len(i['preco'])) % 2 == 0) else (col_3 - len(i['preco']))//2 + 1
        print('|', i['id'],' '*(col_1 - len(i['id'])),'|', ' '*espacos_col_2_esq, i['categoria'], ' '*espacos_col_2_dir, '|', ' '*espacos_col_3_esq, i['preco'], ' '*espacos_col_3_dir, '|')
        print('+' + '-'*(col_1 + col_2 + col_3 + 13) + '+')

def menu(dados):
    '''
    O parâmetro "dados" deve ser uma lista de dicionários representando os produtos.
    Essa função deverá, em loop, realizar as seguintes ações:
    - Exibir as seguintes opções:
        1. Listar categorias
        2. Listar produtos de uma categoria
        3. Produto mais caro por categoria
        4. Produto mais barato por categoria
        5. Top 10 produtos mais caros
        6. Top 10 produtos mais baratos
        0. Sair
    - Ler a opção do usuário.
    - No caso de opção inválida, imprima uma mensagem de erro.
    - No caso das opções 2, 3 ou 4, pedir para o usuário digitar a categoria desejada.
    - Chamar a função adequada para tratar o pedido do usuário e salvar seu retorno.
    - Imprimir o retorno salvo. 
    O loop encerra quando a opção do usuário for 0.
    '''
    rodar_menu = True
    opcoes = ('0', '1', '2', '3', '4', '5', '6')
    while rodar_menu:
        opcao = input('Digite um número para prosseguir:\n 1. Listar categorias\n 2. Listar produtos de uma categoria\n 3. Produto mais caro por categoria\n 4. Produto mais barato por categoria\n 5. Top 10 produtos mais caros\n 6. Top 10 produtos mais baratos\n0. Sair\n ')
        if opcao not in opcoes:
            print('Opção inválida!')
        else:
            if opcao == '1':
                print('As categorias são: \n')
                comp_linha = 0
                for i in range(len(listar_categorias(dados)) - 1):
                    if comp_linha < 50:
                        print(listar_categorias(dados)[i], end = ', ')
                        comp_linha += len(listar_categorias(dados)[i])
                    else:
                        print(listar_categorias(dados)[i] + ',')
                        comp_linha = 0
                print(listar_categorias(dados)[-1]+'.')
            elif opcao == '2':
                categoria = escolher_categoria(dados)
                if categoria == '9':
                    pass
                else:
                    imprimir_dados_tabela(listar_por_categoria(dados, categoria))
            elif opcao == '3':
                categoria = escolher_categoria(dados)
                if categoria == '9':
                    pass
                else:
                    mais_caro = produto_mais_caro(dados, categoria)
                    imprimir_dados_tabela([mais_caro])
                    print(f'Produto mais caro: ID: {mais_caro["id"]}, categoria: {mais_caro["categoria"]} e preço R$ {mais_caro["preco"]}.')
            elif opcao == '4':
                categoria = escolher_categoria(dados)
                if categoria == '9':
                    pass
                else:
                    mais_barato = produto_mais_barato(dados, categoria)
                    imprimir_dados_tabela([mais_barato])
                    print(f'\n Produto mais barato: ID: {mais_barato["id"]}, categoria: {mais_barato["categoria"]} e preço R$ {mais_barato["preco"]}.\n')
            elif opcao == '5':
                imprimir_dados_tabela(top_10_caros(dados))
            elif opcao == '6':
                imprimir_dados_tabela(top_10_baratos(dados))
            elif opcao == '0':
                rodar_menu = False
    else:
        print('Muito obrigado, espero tê-lo ajudado!')


# Programa Principal - não modificar!
dados = obter_dados()
menu(dados)




```

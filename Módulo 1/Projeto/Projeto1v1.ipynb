```python
import json
import os.path
import sys

def obter_dados():
    '''
    Essa função carrega os dados dos produtos e retorna uma lista de dicionários, onde cada dicionário representa um produto.
    NÃO MODIFIQUE essa função.
    '''
    with open(os.path.join(sys.path[0], 'dados.json'), 'r') as arq:
        dados = json.loads(arq.read())
    return dados

def listar_categorias(dados):
    '''
    O parâmetro "dados" deve ser uma lista de dicionários representando os produtos.
    Essa função deverá retornar uma lista contendo todas as categorias dos diferentes produtos.
    Cuidado para não retornar categorias repetidas.    
    '''
    categorias_set = set([])
    for i in dados:
        categorias_set.add(i['categoria'])
    return list(categorias_set)


def listar_por_categoria(dados, categoria):
    '''
    O parâmetro "dados" deve ser uma lista de dicionários representando os produtos.
    O parâmetro "categoria" é uma string contendo o nome de uma categoria.
    Essa função deverá retornar uma lista contendo todos os produtos pertencentes à categoria dada.
    '''
    if categoria in listar_categorias(dados):
        produtos_list = []
        for i in dados:
            if i['categoria'] == categoria:
                produtos_list.append(i)
        return produtos_list
    else: 
        return print('Categoria não encontrada!')         
    

def produto_mais_caro(dados, categoria):
    '''
    O parâmetro "dados" deve ser uma lista de dicionários representando os produtos.
    O parâmetro "categoria" é uma string contendo o nome de uma categoria.
    Essa função deverá retornar um dicionário representando o produto mais caro da categoria dada.
    '''
    produtos_categoria = listar_por_categoria(dados, categoria)
    prod_mais_caro = produtos_categoria[0]
    for i in produtos_categoria:
        if float(i['preco']) > float(prod_mais_caro['preco']):
            prod_mais_caro = i
    return prod_mais_caro


def produto_mais_barato(dados, categoria):
    '''
    O parâmetro "dados" deve ser uma lista de dicionários representando os produtos.
    O parâmetro "categoria" é uma string contendo o nome de uma categoria.
    Essa função deverá retornar um dicionário representando o produto mais caro da categoria dada.
    '''
    produtos_categoria = listar_por_categoria(dados, categoria)
    prod_mais_barato = produtos_categoria[0]
    for i in produtos_categoria:
        if float(i['preco']) < float(prod_mais_barato['preco']):
            prod_mais_barato = i
    return prod_mais_barato

def mais_caro(dados):
    prod_mais_caro = dados[0]   
    for i in dados:
       if float(i['preco']) > float(prod_mais_caro['preco']):
            prod_mais_caro = i
    return prod_mais_caro

def mais_barato(dados):
    prod_mais_barato = dados[0]   
    for i in dados:
       if float(i['preco']) < float(prod_mais_barato['preco']):
            prod_mais_barato = i
    return prod_mais_barato

def top_10_caros(dados):
    '''
    O parâmetro "dados" deve ser uma lista de dicionários representando os produtos.
    Essa função deverá retornar uma lista de dicionários representando os 10 produtos mais caros.
    '''
    top10 = []
    for i in range(10):
        prod_mais_caro = mais_caro(dados)
        top10.append(prod_mais_caro)
        dados.remove(prod_mais_caro)
    return top10
    

def top_10_baratos(dados):
    '''
    O parâmetro "dados" deve ser uma lista de dicionários representando os produtos.
    Essa função deverá retornar uma lista de dicionários representando os 10 produtos mais baratos.
    '''
    top10 = []
    for i in range(10):
        prod_mais_barato = mais_barato(dados)
        top10.append(prod_mais_barato)
        dados.remove(prod_mais_barato)
    return top10
    
def imprimir_dados(dados):
    for i in dados:
        print(f'Produto: ID: {i["id"]}, categoria: {i["categoria"]} e preço R$ {i["preco"]}.')

def menu(dados):
    '''
    O parâmetro "dados" deve ser uma lista de dicionários representando os produtos.
    Essa função deverá, em loop, realizar as seguintes ações:
    - Exibir as seguintes opções:
        1. Listar categorias
        2. Listar produtos de uma categoria
        3. Produto mais caro por categoria
        4. Produto mais barato por categoria
        5. Top 10 produtos mais caros
        6. Top 10 produtos mais baratos
        0. Sair
    - Ler a opção do usuário.
    - No caso de opção inválida, imprima uma mensagem de erro.
    - No caso das opções 2, 3 ou 4, pedir para o usuário digitar a categoria desejada.
    - Chamar a função adequada para tratar o pedido do usuário e salvar seu retorno.
    - Imprimir o retorno salvo. 
    O loop encerra quando a opção do usuário for 0.
    '''
    loop = True
    opcoes = ('0', '1', '2', '3', '4', '5', '6')
    while loop:
        opcao = input('Digite um número para prosseguir:\n 1. Listar categorias\n 2. Listar produtos de uma categoria\n 3. Produto mais caro por categoria\n 4. Produto mais barato por categoria\n 5. Top 10 produtos mais caros\n 6. Top 10 produtos mais baratos\n0. Sair\n ')
        if opcao not in opcoes:
            print('Opção inválida!')
        else:
            if opcao == '1':
                print(listar_categorias(dados))
            elif opcao == '2':
                categoria = input('Digite a categoria que deseja listar: ')
                if categoria not in listar_categorias(dados):
                    print('Categoria não encontrada!')
                else:
                    imprimir_dados(listar_por_categoria(dados, categoria))
            elif opcao == '3':
                categoria = input('Digite a categoria que deseja listar: ')
                if categoria not in listar_categorias(dados):
                    print('Categoria não encontrada!')
                else:
                    mais_caro = produto_mais_caro(dados, categoria)
                    print(f'Produto mais caro: ID: {mais_caro["id"]}, categoria: {mais_caro["categoria"]} e preço R$ {mais_caro["preco"]}.')
            elif opcao == '4':
                categoria = input('Digite a categoria que deseja listar: ')
                if categoria not in listar_categorias(dados):
                    print('Categoria não encontrada!')
                else:
                    mais_barato = produto_mais_caro(dados, categoria)
                    print(f'Produto mais barato: ID: {mais_barato["id"]}, categoria: {mais_barato["categoria"]} e preço R$ {mais_barato["preco"]}.')
            elif opcao == '5':
                imprimir_dados(top_10_caros(dados))
            elif opcao == '6':
                imprimir_dados(top_10_baratos(dados))
            elif opcao == '0':
                loop = False
    else:
        print('Muito obrigado, espero tê-lo ajudado!')


# Programa Principal - não modificar!
dados = obter_dados()
menu(dados)
```
